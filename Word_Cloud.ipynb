```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
bingo = pd.read_csv('/content/drive/MyDrive/DataSets/doubleu_games/bingo.csv')
casino = pd.read_csv('/content/drive/MyDrive/DataSets/doubleu_games/casino.csv')
casinoslots = pd.read_csv('/content/drive/MyDrive/DataSets/doubleu_games/casinoslots.csv')
take5free = pd.read_csv('/content/drive/MyDrive/DataSets/doubleu_games/take5free.csv')
undead = pd.read_csv('/content/drive/MyDrive/DataSets/doubleu_games/undead.csv')
```

```python
bingo_grouped = bingo.groupby(bingo.rating)
bad_bingo = bingo_grouped.get_group(1)
good_bingo = bingo_grouped.get_group(5)

casino_grouped = casino.groupby(casino.rating)
bad_casino = casino_grouped.get_group(1)
good_casino = casino_grouped.get_group(5)

casinoslots_grouped = casinoslots.groupby(casinoslots.rating)
bad_casinoslots = casinoslots_grouped.get_group(1)
good_casinoslots = casinoslots_grouped.get_group(5)

take5free_grouped = take5free.groupby(take5free.rating)
bad_take5free = take5free_grouped.get_group(1)
good_take5free = take5free_grouped.get_group(5)

undead_grouped = undead.groupby(undead.rating)
bad_undead = undead_grouped.get_group(1)
good_undead = undead_grouped.get_group(5)
```

```python
bad_bingo_text = " ".join(cat for cat in bad_bingo.review)
good_bingo_text = " ".join(cat for cat in good_bingo.review)

bad_casino_text = " ".join(cat for cat in bad_casino.review)
good_casino_text = " ".join(cat for cat in good_casino.review)

bad_casinoslots_text = " ".join(cat for cat in bad_casinoslots.review)
good_casinoslots_text = " ".join(cat for cat in good_casinoslots.review)

bad_take5free_text = " ".join(cat for cat in bad_take5free.review)
good_take5free_text = " ".join(cat for cat in good_take5free.review)

bad_undead_text = " ".join(cat for cat in bad_undead.review)
good_undead_text = " ".join(cat for cat in good_undead.review)
```

```python
stop_words = ["game", "play", "bingo", "casino"] + list(STOPWORDS)
```

```python
bad_bingo_word_cloud = WordCloud(width=3000, height=2000, random_state=1, background_color="black", colormap="Pastel1", collocations=False, stopwords=stop_words,).generate(bad_bingo_text)
good_bingo_word_cloud = WordCloud(width=3000, height=2000, random_state=1, background_color="black", colormap="Pastel1", collocations=False, stopwords=stop_words,).generate(good_bingo_text)

bad_casino_word_cloud = WordCloud(width=3000, height=2000, random_state=1, background_color="black", colormap="Pastel1", collocations=False, stopwords=stop_words,).generate(bad_casino_text)
good_casino_word_cloud = WordCloud(width=3000, height=2000, random_state=1, background_color="black", colormap="Pastel1", collocations=False, stopwords=stop_words,).generate(good_casino_text)

bad_casinoslots_word_cloud = WordCloud(width=3000, height=2000, random_state=1, background_color="black", colormap="Pastel1", collocations=False, stopwords=stop_words,).generate(bad_casinoslots_text)
good_casinoslots_word_cloud = WordCloud(width=3000, height=2000, random_state=1, background_color="black", colormap="Pastel1", collocations=False, stopwords=stop_words,).generate(good_casinoslots_text)

bad_take5free_word_cloud = WordCloud(width=3000, height=2000, random_state=1, background_color="black", colormap="Pastel1", collocations=False, stopwords=stop_words,).generate(bad_take5free_text)
good_take5free_word_cloud = WordCloud(width=3000, height=2000, random_state=1, background_color="black", colormap="Pastel1", collocations=False, stopwords=stop_words,).generate(good_take5free_text)

bad_undead_word_cloud = WordCloud(width=3000, height=2000, random_state=1, background_color="black", colormap="Pastel1", collocations=False, stopwords=stop_words,).generate(bad_undead_text)
good_undead_word_cloud = WordCloud(width=3000, height=2000, random_state=1, background_color="black", colormap="Pastel1", collocations=False, stopwords=stop_words,).generate(good_undead_text)
```

```python
fig = plt.figure(figsize=(15,15))
fig.suptitle('Bad Review', size=20)
rows = 5
cols = 1

ax1 = fig.add_subplot(rows, cols, 1)
ax1.imshow(bad_bingo_word_cloud)
ax1.set_title('BIngo Bad Review')
ax1.axis("off")
 
ax2 = fig.add_subplot(rows, cols, 2)
ax2.imshow(bad_casino_word_cloud)
ax2.set_title('Casino Bad Review')
ax2.axis("off")
 
ax3 = fig.add_subplot(rows, cols, 3)
ax3.imshow(bad_casinoslots_word_cloud)
ax3.set_title('Casinoslolts Bad Review')
ax3.axis("off")

ax4 = fig.add_subplot(rows, cols, 4)
ax4.imshow(bad_take5free_word_cloud)
ax4.set_title('take5free Bad Review')
ax4.axis("off")

ax5 = fig.add_subplot(rows, cols, 5)
ax5.imshow(bad_undead_word_cloud)
ax5.set_title('undead Bad Review')
ax5.axis("off")

plt.show()
```

```python
fig = plt.figure(figsize=(15,15))
fig.suptitle('Good Review', size=20)
rows = 5
cols = 1

ax1 = fig.add_subplot(rows, cols, 1)
ax1.imshow(good_bingo_word_cloud)
ax1.set_title('BIngo good Review')
ax1.axis("off")
 
ax2 = fig.add_subplot(rows, cols, 2)
ax2.imshow(good_casino_word_cloud)
ax2.set_title('Casino good Review')
ax2.axis("off")
 
ax3 = fig.add_subplot(rows, cols, 3)
ax3.imshow(good_casinoslots_word_cloud)
ax3.set_title('Casinoslolts good Review')
ax3.axis("off")

ax4 = fig.add_subplot(rows, cols, 4)
ax4.imshow(good_take5free_word_cloud)
ax4.set_title('take5free good Review')
ax4.axis("off")

ax5 = fig.add_subplot(rows, cols, 5)
ax5.imshow(good_undead_word_cloud)
ax5.set_title('undead good Review')
ax5.axis("off")

plt.show()
```
